```python
# 1 way

try:

    num1 = int(input("num1:"))  
    num2 = int(input("num2:")) 
    res = num1/num2
    
except ValueError as e:
    print('some error please try again letter')
    print(e)
    
except ZeroDivisionError as e:
    
    print(e)

except:
    print('Error')

else:
     print(res)
        
finally:
    print('hello')

print('trasaction completed')
```

```python
# 2 way

try:

    num1 = int(input("num1:"))  
    num2 = int(input("num2:")) 
    res = num1/num2
    
except Exception as e:
    print(e)

else:
     print(res)
        
finally:
    print('hello')

print('trasaction completed')
```

```python
# nested try and except block

try:
    a = eval(input('enter the 1st number'))  # 'abc'
    try:
        b = int(input('enetr the second number')) #0
        try:
            print(a/b)
        except Exception as e:
            print('line 10' , e)
    except Exception as e:
        print('line 12' , e)
except Exception as e:
    print('line 14' , e)        

print('hello its done')
```

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
```

```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
import re
```

```python
pattern = re.compile(r'khanafsaan11.com') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# dot(.) - any character except new line

pattern = re.compile(r'.') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# \d - digit (0-9)

pattern = re.compile(r'\d') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# \D - not digit (0-9)

pattern = re.compile(r'\D') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# \w - word char (a-z , A-Z , 0-9 , _)

pattern = re.compile(r'\w') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# \W - not word char (a-z , A-Z , 0-9 , _)

pattern = re.compile(r'\W') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# \s - whitsapce 

pattern = re.compile(r'\s') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```

```python
# \S - not whitsapce 

pattern = re.compile(r'\S') #  whta you want to search for
matches = pattern.finditer(text_to_search)


for match in matches:
    print(match)
```
